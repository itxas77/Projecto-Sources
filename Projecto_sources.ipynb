#  Introducción al tema


La música rock ha sido un género dominante durante décadas, influyendo en la cultura, la moda y la industria musical.  
En esta presentación breve, exploraremos cómo ha evolucionado su presencia en las plataformas de streaming y el consumo global de música.

# Planteamiento de la hipótesis

**Hipótesis:**  

> "El rock ha perdido relevancia en el consumo musical global y ha sido desplazado por otros géneros en las plataformas de streaming."

#  Datasets utilizados

Para contrastar esta hipótesis, analizaremos tres datasets clave:

1. **`history_of_rock_in_spotify.csv`**  
   Contiene datos históricos de reproducciones de canciones de rock en Spotify, mostrando tendencias de popularidad a lo largo del tiempo.

2. **`Global_Music_Streaming_Listener_Preferences.csv`**  
   Muestra las preferencias de los oyentes de música a nivel global, por género y región.

3. **`Spotify_2024_Global_Streaming_Data.csv`**  
   Datos recientes de streaming global, incluyendo reproducciones por género, país y artista.

#  Explicación de la metodología

Vamos a:

- Examinar la evolución del rock a lo largo de los años en Spotify.
- Comparar la popularidad del rock frente a otros géneros según las preferencias de los oyentes.
- Analizar los datos recientes para ver la posición del rock en el consumo global de música en 2024.
